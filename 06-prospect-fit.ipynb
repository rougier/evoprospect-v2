### Fit gain and loss together

```python
excluded_monkeys = ['ANU', 'NER', 'YIN', 'OLG', 'JEA', 'PAT', 'YOH', 'GAN'] # REMOVE GAN ???


with open('results-fits/monkey-analysis-L0-params.pkl', 'rb') as file:
    params = pickle.load(file)
fits = {}
players = list(params.values())[0] # Name of the players
for player in players:
    fits[player.shortname] = {}
for monkey in params.keys():
    # rejected monkeys
    if monkey not in excluded_monkeys:
        for player_fit in params[monkey]:
            fits[player_fit.shortname][monkey] = player_fit.parameters

for model in ['P1', 'TK',  'P2',  'GE', 'TK']:
    if model in fits:
        fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))
        fig.suptitle('{} fit'.format(model))
        plot_sigmoid(ax1, fits[model], compute_mean_fit(fits[model]))
        plot_utility(ax2, fits[model], compute_mean_fit(fits[model]))
        plot_probability(ax3, fits[model], compute_mean_fit(fits[model]), type=model, domain=None)
        plt.show()


 
```

### Fit gain and loss separately

```python
excluded_monkeys = ['ANU', 'NER', 'YIN', 'OLG', 'JEA', 'PAT', 'YOH', 'GAN'] # REMOVE GAN ???


with open('results-fits/monkey-analysis-L0-params.pkl', 'rb') as file:
    params = pickle.load(file)
fits = {}
players = list(params.values())[0] # Name of the players
for player in players:
    fits[player.shortname] = {}
for monkey in params.keys():
    # rejected monkeys
    if monkey not in excluded_monkeys:
        for player_fit in params[monkey]:
            fits[player_fit.shortname][monkey] = player_fit.parameters

for model in ['P1+', 'TK+', 'P2+', 'GE+']:
    if model in fits:
        fig, axes = plt.subplots(2, 2, figsize=(12, 10))
        fig.suptitle(f'{model} fit')

        plot_sigmoid(axes[0, 0], fits[model], compute_mean_fit(fits[model]))
        plot_utility(axes[0, 1], fits[model], compute_mean_fit(fits[model]))
        plot_probability(axes[1, 0], fits[model], compute_mean_fit(fits[model]), type=model, domain='gain')
        plot_probability(axes[1, 1], fits[model], compute_mean_fit(fits[model]), type=model, domain='loss')

        plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout and title spacing
        plt.show()
    

 
```
